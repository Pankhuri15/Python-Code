<h1> Careers with lowest risk vs. most potential.

<h3> Careers with lowest risk.

```python
spread_col = clean_df["Mid-Career 90th Percentile Salary"] - clean_df["Mid-Career 10th Percentile Salary"]
```

```python
clean_df.insert(1, "Spread", spread_col)
```

```python
clean_df.head()
```

```python
lowest_risk = clean_df.sort_values("Spread")
```

```python
lowest_risk[["Undergraduate Major", "Spread"]].head()
```

<h3> Careers with highest potential.

```python
highest_potential = clean_df.sort_values("Mid-Career 90th Percentile Salary", ascending=False)
```

```python
highest_potential[["Undergraduate Major", "Mid-Career 90th Percentile Salary"]].head()
```

<h3> Degrees with greatest spread.

```python
lowest_risk[["Undergraduate Major", "Spread"]].tail()
```

<h3> Highest median salary careers.

```python
highest_median = clean_df.sort_values("Mid-Career Median Salary", ascending=False)
highest_median[["Undergraduate Major", "Mid-Career Median Salary"]].head()
```
